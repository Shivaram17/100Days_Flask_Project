```python
# leet code Array and remove element
class Solution:
    def removeElement(self, nums, val):
        l = len(nums)
        count = 0
        for i in range(l):
            if nums[i] != val:
                nums[count] = nums[i]
                count +=1
        # c = nums[:k]
        return count

nums = [3,2,2,3]
val = 3
obj = Solution()
obj.removeElement(nums, val)
```

```python
# Remove duplicates from a sorted array without extra array
class Solution:
    def removeDuplicates(self, nums):
        k = 0
        for i in range(len(nums)):
            if i < len(nums) -1 and nums[i] == nums[i+1]:
                continue
            nums[k] = nums[i]
            k +=1
        return k, nums
nums = [1,1,3]
obj = Solution()
obj.removeDuplicates(nums)
```
